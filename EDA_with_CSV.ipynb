## EDA with CSV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/employee.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
rows, cols = df.shape
print(f'Rows: {rows}')
print(f'Cols: {cols}')
```

```python
df.info()
```

```python
# review the statastic
df.describe()
```

```python
print('Missing Values:')
df.isnull().sum()
```

```python
print(f'Duplicate Records: {df.duplicated().sum()}')
```

```python
# Taking Care of Missing Values
df['Name'] = df['Name'].fillna('Unknown')
df
```

```python
print('Age Mean', df['Age'].mean())
print('Age Median', df['Age'].median())
print('Age Mode', df['Age'].mode())
```

```python
age_median = df['Age'].median()
df['Age'] = df['Age'].fillna(age_median)    
```

```python
print('Weight Mean', df['Weight'].mean())
print('Weight Median', df['Weight'].median())
print('Weight Mode', df['Weight'].mode())
```

```python
weight_median = df['Weight'].median()
df['Weight'] = df['Weight'].fillna(weight_median)
df
```

```python
print('Salary Mean', df['Salary'].mean())
print('Salary Median', df['Salary'].median())
print('Salary Mode', df['Salary'].mode())

salary_median = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(salary_median)
df
```

```python
print('Missing Values:')
df.isnull().sum()
```

```python
df = df.sort_values(by='Age', ascending=True)

plt.figure(figsize=(8,5))
plt.plot(df['Age'], df['Salary'], marker='o')
plt.xlabel('Age')
plt.ylabel('Salary (In Diamonds)')
plt.title('Line Graph of Salary by Age')
plt.grid(linestyle=':')
plt.show()
```
