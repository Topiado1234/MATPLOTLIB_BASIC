## Data Visualization with Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Basic Lines Plots

```python
 # Creating data
xpoint = np.array([1, 8])
ypoint = np.array([3, 10])

xpoint
ypoint
```

```python
# State oriented Approach. 1st stage(get good here 1st)
plt.plot(xpoint, ypoint)
plt.title('RTT-13 First Plot')
plt.xlabel('X- Axis')
plt.ylabel('Y- Axis')
plt.show() # Show our plot even if in a .py file
```

```python
# Object oriented Approach. Little more control
fig, ax = plt.subplots()
ax.plot(xpoint, ypoint, marker='d')
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 170, 190, 160, 300, 310, 320, 330])
```

```python
plt.plot(x, y, marker='D', linestyle=':', color='blue')
plt.title('Title: Sports Watch Data')
plt.xlabel('Average Pulse Rate')
plt.xticks(np.arange(80, 140, 5))
plt.ylabel('Calorie Burnage')
plt.show()
```

```python
# RGB
plt.plot(x, y, marker='D', linestyle=':', color='0.1')
plt.title('Title: Sports Watch Data')
plt.xlabel('Average Pulse Rate')
plt.xticks(np.arange(80, 140, 5))
plt.ylabel('Calorie Burnage')
plt.show()
```

```python
plt.plot(x, y, marker='D', linestyle=':', color='#FF5733')
plt.title('Title: Sports Watch Data', color='red')
plt.xlabel('Average Pulse Rate', color='#AA6677')
plt.xticks(np.arange(80, 140, 5), color= 'green')
plt.ylabel('Calorie Burnage')
plt.show()
```

```python
years=[1970,1980,1985,1990,1995,2000,2005,2010,2015,2020,2025,2030,2035]
pops=[3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3,7.3,7.3,7.4,7.5]
```

```python
plt.plot(facecolor='lightred')
plt.plot(years, pops, marker='D', linestyle=':', color='#FF5733')
plt.title('World Population Graph')
plt.xlabel('Population Growth by Year', color = 'green')
plt.ylabel('Population in Billions', color= 'red')
```

## Font Styles
- fontsize - size of text
- fontfamily

```python
# coloring line with kwargs
plt.plot(x, y, marker='D', linestyle=':', color='blue')
plt.title('Title: Sports Watch Data', fontweight='bold', fontsize=30)
plt.xlabel('Average Pulse Rate', fontsize=20, fontweight='bold')
plt.xticks(np.arange(80, 140, 5))
plt.ylabel('Calorie Burnage', fontfamily='monospace', fontweight='bold')
plt.show()
```

```python
# coloring line with kwargs - key work argument
# pre-created font class
label_font = {
    'family': 'sans-serif',
    'color':'green',
    'weight': 'normal',
    'size':20
}

plt.plot(x, y, marker='D', linestyle=':', color='blue')
plt.title('Title: Sports Watch Data')
plt.xlabel('Average Pulse Rate', fontdict=label_font)
plt.xticks(np.arange(80, 130, 5))
plt.ylabel('Calorie Burnage', fontdict=label_font)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10,4,200,45,67,56,98,100,])
plt.plot(ypoints, marker='D', linestyle='--')
plt.grid(color='red',linestyle='--')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10,4,200,45,67,56,98,100,])
plt.plot(ypoints, marker='D', linestyle='--', alpha=0.25)
plt.show()
```

```python
plt.plot(x, y, marker='D', linestyle=':', color='blue')

plt.title('Title: Sports Watch Data')
plt.xlabel('Average Pulse Rate')
plt.xticks(np.arange(80, 130, 5))
plt.ylabel('Calorie Burnage')
plt.grid(color='red')
plt.gca().set_axisbelow(True)
plt.gca().set_facecolor('lightyellow') #

plt.show()
```

```python
# coloring line with kwargs - key work argument
plt.plot(x, y, marker='D', linestyle=':', color='blue')

plt.title('Sports Watch Data')
plt.xlabel('Average Pulse Rate')
plt.xticks(np.arange(80, 130, 5))
plt.ylabel('Calorie Burnage')

plt.grid(color='red')
plt.gca().set_axisbelow(True) # set gridlines behind plot elements

plt.gca().set_facecolor('lightyellow') # set canvas background color

plt.show()
```

### Multi line  plots

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]


plt.plot(years, pops, color='green', linestyle='--', marker='d', label='Population')
plt.plot(years, death, color='red', linestyle='dashdot', marker='o', label='Death')

plt.title('Population Growth')
plt.xlabel('Year')
plt.ylabel('Population Growth by Year (Billions)')

plt.show()
```

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
pops=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
death=[1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]
job_rate = [0.5, 0.6, 0.8, 0.9, 1.0, 1.2, 1.5, 1.7, 2.0, 2.2, 2.5, 2.7, 2.9]

plt.figure(figsize=(10, 8))

plt.plot(years, pops, color='green', linestyle='--', marker='d', label='Population')
plt.plot(years, death, color='red', linestyle='dashdot', marker='o', label='Death')
plt.plot(years, job_rate, color='blue', linestyle='dotted', marker='s', label='Job Rate')
plt.grid(linestyle='dotted')

plt.title('Population Growth')
plt.xlabel('Year')
plt.ylabel('Population Growth by Year (Billions)')

plt.legend()

plt.show()
```
